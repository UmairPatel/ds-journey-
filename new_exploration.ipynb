Titanic Dataset Analysis and Interpretation. 
In this notebook I loaded a CSV file to interpret a data set as well as answered important questions such as survival rates and how likely people were to survive depending on their gender, wealth, etc. I also found missing values and added them up. 

```python
import pandas as pd

df = pd.read_csv('train.csv')

print(df.head(5))
print(df.tail(5))
print(df.info())
print(df.shape)
```

```python
print(df['Survived'].value_counts())
print(df['Age'].mean())
print(df['Fare'].mean())
```

```python
print(df['Sex'].value_counts())
print(df.groupby('Sex')['Survived'].mean())
print(df.groupby('Pclass')['Survived'].mean())
```

```python
print(df['Age'].max())
print(df['Age'].min())
print(df['Age'].isnull().sum())
print(f"The most expensive fare was: ${df['Fare'].max():,.2f}")
```

```python
#Task 1: Bar Chart showing passengers who survived vs did not survive.

import matplotlib.pyplot as plt
import seaborn as sns

sns.countplot(data=df, x="Survived")
plt.title('Passengers who survived vs who did not survive')
plt.xlabel('Passengers')
plt.ylabel('Amount of Passengers')
plt.xticks([0,1], ['Passengers who did not Survive','Passengers who survived'])
plt.show()
```

```python
#Task 2: Survival Rate by Gender

sns.countplot(data=df, x = "Sex", hue="Survived")
plt.title('Survival Rate by Gender')
plt.xlabel('Gender')
plt.ylabel('Survival Rate')
plt.legend(title='Survived', labels=['Did not Survive', 'Survived'])
plt.show()
```

```python
#Task 3: Survival Rate by Passenger Class. 

sns.countplot(data=df, x = "Pclass", hue='Survived')
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Amount of Passenger')
plt.legend(title='Survived', labels=['Did not Survive', 'Survived'])
plt.show()
```

```python
#Task 4: Distribution of passenger Ages

sns.histplot(data=df, x="Age")
plt.title('Distribution of Passenger Ages')
plt.xlabel('Age')
plt.ylabel('Amount of Passengers')
plt.show()
```

```python
#Task 5: Fair distribution by Passenger Class

sns.boxplot(data=df, y="Fare", x="Pclass")
plt.title('Fare Distribution by Passenger class')
plt.xlabel('Passenger Class')
plt.ylabel('Fare Amount in Dollars')
plt.xticks([0, 1, 2], labels=['1st Class', '2nd Class', '3rd Class'])
plt.show()
```
